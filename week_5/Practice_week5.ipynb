# 🤔 Lesson 1 — AI vs Traditional Programming

- **Traditional program:** YOU write every rule (`if` / `else`).
- **AI / Machine Learning:** gives the computer **labelled examples** and it learns the pattern.
- **Inference:** using a trained model to predict on **new** data it has never seen.

````python
# Traditional approach: YOU write the rules
def is_spam_rules(message):
    message = message.lower()
    if "win a prize" in message:
        return True
    if "click here" in message:
        return True
    if "free money" in message:
        return True
    return False

messages = [
    "You win a prize! Click here!",
    "Hey, are you free tomorrow?",
    "Free money – claim now!",
    "Study session tonight?",
]

for msg in messages:
    label = "🚫 SPAM" if is_spam_rules(msg) else "✅ Not spam"
    print(label, "→", msg)
````

````python
# AI approach: LEARN from labelled examples

# Step 1 — collect labelled data
training = [
    ("win a prize free money",    "spam"),
    ("click here claim reward",   "spam"),
    ("congratulations you won",   "spam"),
    ("study session at 6pm",      "not spam"),
    ("are you free tomorrow",     "not spam"),
    ("meeting cancelled today",   "not spam"),
    ("limited offer free gift",   "spam"),
    ("see you at school",         "not spam"),
]

# Step 2 — train: count how often each word appears in spam vs not-spam
spam_count  = {}
normal_count = {}

for text, label in training:
    for word in text.split():
        if label == "spam":
            spam_count[word] = spam_count.get(word, 0) + 1
        else:
            normal_count[word] = normal_count.get(word, 0) + 1

print("✅ Model trained on", len(training), "examples")
print("Top spam words:  ", sorted(spam_count, key=spam_count.get, reverse=True)[:5])
print("Top normal words:", sorted(normal_count, key=normal_count.get, reverse=True)[:5])
````

````python
# Step 3 — inference: use the model on new messages
def predict_spam(message):
    words = message.lower().split()
    spam_score   = sum(spam_count.get(w, 0)   for w in words)
    normal_score = sum(normal_count.get(w, 0) for w in words)
    if spam_score > normal_score:
        return "🚫 SPAM"
    if normal_score > spam_score:
        return "✅ Not spam"
    return "🤷 Unsure"

test_messages = [
    "Win a free prize now",
    "Can we study together",
    "Claim your free gift today",
    "Football practice tomorrow",
]

for msg in test_messages:
    print(predict_spam(msg), "→", msg)
````

# 👁️ Lesson 2 — AI Senses: CV, NLP, Robotics

| Sense | What AI reads | Real example |
|-------|--------------|------|
| **Computer Vision (CV)** | Images = grids of numbers | Face ID, filters |
| **NLP** | Text / speech = word tokens | Siri, Translate |
| **Robotics** | Sensors → decisions → actions | Self-driving car |

````python
import numpy as np
import matplotlib.pyplot as plt

# Computer Vision: an image is just a grid of numbers
# 0 = black  |  255 = white
pixel_grid = np.array([
    [  0,   0, 255, 255,   0,   0],
    [  0, 255, 255, 255, 255,   0],
    [255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255],
    [  0, 255, 255, 255, 255,   0],
    [  0,   0, 255, 255,   0,   0],
], dtype=np.uint8)

print("Image shape:", pixel_grid.shape, "← (rows, columns)")
print("Top row as numbers:", pixel_grid[0])

plt.imshow(pixel_grid, cmap="gray", vmin=0, vmax=255)
plt.title("A 6×6 image — just numbers!")
plt.axis("off")
plt.show()
````

````python
# NLP: text is split into tokens, then mapped to numbers

sentence = "Python is the best programming language"

# Tokenise
tokens = sentence.lower().split()
print("Tokens:", tokens)

# Build a vocabulary (word → number ID)
vocab = {word: i for i, word in enumerate(tokens)}
print("Vocabulary:", vocab)

# Encode the sentence as a list of numbers
encoded = [vocab[word] for word in tokens]
print("Encoded:  ", encoded)
````

````python
# Robotics: sense → decide → act

def robot_decision(distance_cm, battery_pct):
    if battery_pct < 10:
        return "🔴 Return to base (low battery)"
    if distance_cm < 20:
        return "🛑 Stop  (obstacle too close)"
    if distance_cm < 60:
        return "🐢 Slow down"
    return "🟢 Move forward"

readings = [
    (100, 80),   # (distance cm, battery %)
    ( 45, 60),
    ( 15, 40),
    ( 80,  5),
]

for dist, batt in readings:
    action = robot_decision(dist, batt)
    print(f"Distance: {dist:3d}cm  Battery: {batt:2d}%  →  {action}")
````

# 🔄 Lesson 3 — The ML Cycle: Collect → Train → Predict

```
collect labelled data  →  train model  →  predict on new data
```

| Step | What happens |
|------|--------------|
| **Collect** | Gather examples with correct labels |
| **Train** | Model learns patterns from the examples |
| **Predict** | Use the model on new, unseen data |

````python
# Step 1 — collect labelled movie reviews
training_data = [
    ("I love this movie",           "positive"),
    ("Best film ever",              "positive"),
    ("What an amazing story",       "positive"),
    ("Great acting and plot",       "positive"),
    ("Horrible and boring film",    "negative"),
    ("I hated every minute",        "negative"),
    ("Worst movie I ever saw",      "negative"),
    ("Terrible waste of time",      "negative"),
]

# Step 2 — train: count word frequencies per label
pos_words = {}
neg_words = {}

for text, label in training_data:
    for word in text.lower().split():
        if label == "positive":
            pos_words[word] = pos_words.get(word, 0) + 1
        else:
            neg_words[word] = neg_words.get(word, 0) + 1

print("✅ Trained on", len(training_data), "reviews")
print("Positive words:", sorted(pos_words, key=pos_words.get, reverse=True)[:6])
print("Negative words:", sorted(neg_words, key=neg_words.get, reverse=True)[:6])
````

````python
# Step 3 — predict on new reviews
def predict_review(text):
    words = text.lower().split()
    pos_score = sum(pos_words.get(w, 0) for w in words)
    neg_score = sum(neg_words.get(w, 0) for w in words)
    if pos_score > neg_score:
        return "😊 Positive"
    if neg_score > pos_score:
        return "😞 Negative"
    return "🤷 Unsure"

reviews = [
    "Amazing and great movie",
    "Boring and terrible acting",
    "I love the story",
    "Worst film ever",
]

for r in reviews:
    print(predict_review(r), "→", r)
````

# 🏆 Final Challenge — Color Mood Classifier

Build an AI that reads a pixel grid, finds the **dominant colour channel** (R / G / B), and predicts the **mood**.

````python
import numpy as np
import matplotlib.pyplot as plt

# A 4×4 mostly-red image  (each pixel = [R, G, B])
image = np.array([
    [[220, 50, 50], [210, 45, 55], [230, 55, 45], [215, 50, 50]],
    [[200, 60, 60], [225, 48, 48], [218, 52, 52], [222, 50, 50]],
    [[215, 55, 55], [212, 47, 53], [220, 50, 50], [210, 60, 45]],
    [[225, 45, 55], [218, 50, 50], [215, 55, 50], [220, 50, 48]],
], dtype=np.uint8)

# Step 1: average each colour channel across all pixels
avg_r = image[:, :, 0].mean()
avg_g = image[:, :, 1].mean()
avg_b = image[:, :, 2].mean()
print(f"Average colour — R: {avg_r:.0f}  G: {avg_g:.0f}  B: {avg_b:.0f}")

# Step 2: classify mood by dominant channel
def classify_mood(r, g, b):
    dominant = max([("Red", r), ("Green", g), ("Blue", b)], key=lambda x: x[1])
    moods = {
        "Red":   "😡 Energetic / Passionate",
        "Green": "🌿 Calm / Natural",
        "Blue":  "💙 Peaceful / Sad",
    }
    return dominant[0], moods[dominant[0]]

channel, mood = classify_mood(avg_r, avg_g, avg_b)
print(f"Dominant channel: {channel}")
print(f"Predicted mood:   {mood}")

# Step 3: show the image
plt.imshow(image)
plt.title(f"{channel} → {mood}")
plt.axis("off")
plt.show()
````

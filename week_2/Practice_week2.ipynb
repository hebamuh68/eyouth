#  Part 1 — Operators (The Brain of Python)


## 1.1 Logical Operators 🧠
Used to combine conditions.

- `and` → both must be True
- `or` → at least one True
- `not` → flips True/False


```python

x = 10
y = 5

print(x > 5 and y < 10)   # True and True → True
print(x > 20 or y < 10)   # False or True → True
print(not(x > 5))         # not True → False
```

## 1.2 Assignment Operators ✏️
Shortcut ways to update values.

```python

x = 10
x += 5   # x = x + 5
x -= 3   # x = x - 3
x *= 2   # x = x * 2
print(x)
```

## 1.3 Comparison Operators ⚖️

```python

print(5 == 5)
print(5 != 3)
print(10 > 7)
print(4 < 2)
```

## 1.4 Membership Operators 🔍
Check if something exists.

```python

animals = ["cat", "dog", "lion"]

print("cat" in animals)
print("bird" not in animals)
```

# 💬 Part 2 — Strings (Text Like Apps)


## 2.1 Basic String Input 🎤

```python

animal = input("Enter your favorite animal: ")
print(f"I also love {animal}!")
```

## 2.2 String Properties 📏

```python

text = "Python"
print(len(text))   # length
print(text[0])     # first letter
print(text[-1])    # last letter
```

## 2.3 Formatting Strings 🎯

```python

name = "Heba"
age = 16

print(f"My name is {name} and I am {age} years old")
```

## 2.4 String Transformations 🔄

```python

text = "hello world"

print(text.upper())
print(text.lower())
print(text.replace("world", "Python"))
```

# 🧭 Part 3 — Conditionals (Decision Making)


## 3.1 if / elif / else 🔀
Used to make decisions.

```python

age = 15

if age < 13:
    print("Child")
elif age < 18:
    print("Teenager")
else:
    print("Adult")
```

## 3.2 Nested if 🧩
An if inside another if.

```python

age = 16
has_id = True

if age >= 16:
    if has_id:
        print("Allowed in 🎟️")
    else:
        print("Need ID ❌")
else:
    print("Too young 🚫")
```

# 🎮 Mini Project — Game Access System



You are building a simple game login system:

Rules:
- Must be 13 or older
- Must have a username
- If both true → access granted


```python

age = int(input("Enter your age: "))
username = input("Enter username: ")

if age >= 13 and username != "":
    print("✅ Access granted to the game!")
else:
    print("❌ Access denied")
```

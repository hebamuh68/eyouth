## 3 — Activity: Rock, Paper, Scissors scorekeeper (core logic)

```python
def winner(a, b):
    """Return 'tie', 'a', or 'b' for symbols R/P/S."""
    order = "RPS"
    if a == b:
        return "tie"
    pair = a + b
    if pair in ("RS", "SP", "PR"):
        return "a"
    return "b"

print(winner("R", "S"), winner("S", "R"), winner("P", "P"))
```

```python
# Fixed sequence of rounds for testing (replace with input() later)
rounds = [("R", "S"), ("P", "R"), ("S", "S"), ("R", "P")]
score_a = 0
score_b = 0
for ra, rb in rounds:
    w = winner(ra, rb)
    if w == "a":
        score_a += 1
    elif w == "b":
        score_b += 1
    print(ra, rb, "->", w, "|", score_a, score_b)
print("Final:", score_a, score_b)
```

## 4 — Arcade RPS: walkthrough scaffold (menu + rounds)

```python
def normalize_choice(text):
    t = text.strip().upper()
    if t in ("R", "ROCK"):
        return "R"
    if t in ("P", "PAPER"):
        return "P"
    if t in ("S", "SCISSORS"):
        return "S"
    return ""

print(normalize_choice(" rock "), normalize_choice("paper"), normalize_choice("x"))
```

```python
import random

def bot_play():
    return random.choice("RPS")

random.seed(0)
for _ in range(5):
    print(bot_play(), end=" ")
print()
```

## 7 — Error handling: `try` / `except`

```python
def to_int_safe(text):
    try:
        return int(text)
    except ValueError:
        return None

print(to_int_safe("42"), to_int_safe("oops"))
```

```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "undefined"

print(divide(10, 2), divide(10, 0))
```

```python
data = ["1", "2", "x", "4"]
clean = []
for item in data:
    try:
        clean.append(int(item))
    except ValueError:
        pass
print(clean)
```

## 8 — Activity: secure login and gatekeeper

```python
USERS = {"admin": "secret12", "guest": "hello"}

def login_ok(user, pwd):
    return USERS.get(user) == pwd

print(login_ok("admin", "secret12"))
print(login_ok("admin", "wrong"))
```

```python
def gate(action):
    if action == "read":
        return True
    if action == "write":
        return False
    return False

for act in ["read", "write", "delete"]:
    print(act, gate(act))
```

## 9 — Activity: secure vault login (walkthrough)

```python
VAULT_PIN = "5924"

def vault_access(pin):
    if pin == VAULT_PIN:
        return "unlocked"
    return "locked"

print(vault_access("5924"), vault_access("0000"))
```

```python
def vault_session(pin, blocked):
    if blocked:
        return "blocked"
    try:
        if int(pin) < 0:
            return "invalid"
    except ValueError:
        return "invalid"
    return vault_access(pin)

print(vault_session("5924", False))
print(vault_session("abcd", False))
```

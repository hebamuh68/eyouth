#  Lesson 1 — Introduction to Lists (Ordered Data)

## What is a List?

A **list** is like a numbered shopping list or a playlist — it keeps things **in order**, and you can look up any item by its **position** (called an *index*).

```
Index →    0          1          2          3
       ["Blinding", "Levitating", "Stay", "Peaches"]
```

> 🚨 **Important:** Python counts from **0**, not 1!

### Why use lists?
- Store multiple values in **one variable**
- Keep things in **order**
- **Add**, **remove**, or **change** items easily

````python
# 🎵 Creating your first list — a music playlist
playlist = ["Blinding Lights", "Levitating", "Stay", "Peaches", "Industry Baby"]

print("My playlist:", playlist)
print("First song:", playlist[0])   # index 0 = first item
print("Last song:", playlist[-1])   # -1 is a shortcut for the last item!
print("Number of songs:", len(playlist))
````

````python
# ✏️ Modifying a list

# Add a song to the END
playlist.append("Heat Waves")
print("After append:", playlist)

# Insert at a specific position (index 2)
playlist.insert(2, "As It Was")
print("After insert:", playlist)

# Remove a specific song
playlist.remove("Stay")
print("After remove:", playlist)

# Change a song
playlist[0] = "Save Your Tears"
print("After change:", playlist)
````

````python
# ✂️ Slicing — grab a chunk of the list
# Syntax: list[start:end]  (end is NOT included)

top3 = playlist[0:3]   # first 3 songs
print("Top 3:", top3)

last2 = playlist[-2:]  # last 2 songs
print("Last 2:", last2)

# Check if a song is IN the playlist
print("Is 'Peaches' in playlist?", "Peaches" in playlist)
print("Is 'Shape of You' in playlist?", "Shape of You" in playlist)
````

````python
# 📊 Sorting & other useful operations
scores = [78, 95, 61, 88, 72, 100, 55]

print("Original:", scores)
print("Sorted:", sorted(scores))           # doesn't change original
print("Reversed:", sorted(scores, reverse=True))
print("Highest score:", max(scores))
print("Lowest score:", min(scores))
print("Average:", sum(scores) / len(scores))
````

# 📖 Lesson 2 — Introduction to Dictionaries

## What is a Dictionary?

A **dictionary** stores data as **key → value** pairs. Think of a real dictionary: you look up a **word** (key) and get its **definition** (value).

Or think of a **contact in your phone**: the *name* is the key, and the *phone number* is the value.

```
{
  "name": "Ahmed",
  "age": 15,
  "city": "Cairo"
}
```

### List vs Dictionary
| Feature | List | Dictionary |
|---------|------|------------|
| Access by | Position (index) | Name (key) |
| Order | Yes | Yes (Python 3.7+) |
| Best for | Sequences | Labelled data |

````python
# 👤 Creating a dictionary — a student profile
student = {
    "name": "Sara",
    "age": 15,
    "grade": "9th",
    "favourite_subject": "Computer Science",
    "scores": [88, 92, 75, 95]  # values can be lists too!
}

print("Name:", student["name"])
print("Age:", student["age"])
print("Scores:", student["scores"])
print("Best score:", max(student["scores"]))
````

````python
# ✏️ Modifying a dictionary

# Add a new key-value pair
student["city"] = "Cairo"
print("After adding city:", student)

# Update an existing value
student["age"] = 16
print("Updated age:", student["age"])

# Delete a key
del student["grade"]
print("After deleting grade:", student)

# Safe access — use .get() to avoid errors if key doesn't exist
print("School:", student.get("school", "Not specified"))  # returns default if missing
````

````python
# 🔑 Exploring keys, values, and items
game = {
    "title": "Minecraft",
    "genre": "Sandbox",
    "rating": 9.5,
    "players": "Multiplayer",
    "year": 2011
}

print("Keys:", list(game.keys()))
print("Values:", list(game.values()))
print("\nAll info:")
for key, value in game.items():
    print(f"  {key}: {value}")
````

````python
# 🏠 Nested dictionaries — a contact book
contacts = {
    "Ahmed": {"phone": "010-1234567", "email": "ahmed@mail.com"},
    "Sara":  {"phone": "011-7654321", "email": "sara@mail.com"},
    "Mia":   {"phone": "012-9876543", "email": "mia@mail.com"}
}

# Access nested data
print("Ahmed's phone:", contacts["Ahmed"]["phone"])
print("Sara's email:", contacts["Sara"]["email"])

# Add a new contact
contacts["Omar"] = {"phone": "015-1112233", "email": "omar@mail.com"}
print("\nAll contacts:")
for name, info in contacts.items():
    print(f"  {name} → {info['phone']}")
````

# 🔵 Lesson 3 — Sets: Handling Unique Data

## What is a Set?

A **set** is like a bag where **every item can only appear once** — duplicates are automatically removed. Sets also have no guaranteed order.

Think of it like a **guest list** — you don't want the same person listed twice!

```
my_set = {1, 2, 3, 4, 5}   ← no duplicates!
```

### When to use sets?
- Remove duplicates from data
- Check if something is in a group (very fast!)
- Compare two groups (who's in both? who's only in one?)

````python
# 🎉 Sets in action — unique party guests
guest_list = {"Ahmed", "Sara", "Mia", "Omar", "Sara", "Ahmed"}  # duplicates!
print("Unique guests:", guest_list)  # Sara and Ahmed only appear once
print("Guest count:", len(guest_list))

# Add a new guest
guest_list.add("Layla")
print("After adding Layla:", guest_list)

# Remove a guest
guest_list.discard("Mia")  # discard won't crash if item doesn't exist
print("After removing Mia:", guest_list)

# Check membership (very fast!)
print("Is Omar invited?", "Omar" in guest_list)
````

````python
# 🔢 Set Math — comparing groups
class_a = {"Ahmed", "Sara", "Mia", "Omar"}
class_b = {"Mia", "Layla", "Karim", "Omar"}

# Union — everyone from both classes
print("All students:", class_a | class_b)

# Intersection — students in BOTH classes
print("In both classes:", class_a & class_b)

# Difference — in class_a but NOT class_b
print("Only in Class A:", class_a - class_b)

# Symmetric difference — in one class but NOT both
print("In exactly one class:", class_a ^ class_b)
````

````python
# 🧹 Real-world use: removing duplicates from a list
votes = ["Minecraft", "Roblox", "Minecraft", "FIFA", "Roblox", "Minecraft", "GTA V"]
print("All votes:", votes)
print("Total votes:", len(votes))

# Get unique games
unique_games = set(votes)
print("Unique games:", unique_games)

# Convert back to sorted list
unique_list = sorted(list(unique_games))
print("Sorted unique games:", unique_list)
````

# 🔄 Lesson 4 — Iterating Over Collections

## What is Iteration?

**Iterating** means going through each item in a collection, one at a time. You already know `for` loops — now let's use them with lists, dictionaries, and sets!

```python
for item in my_collection:
    # do something with each item
```

````python
# 📋 Iterating over a list
subjects = ["Maths", "Science", "English", "Computer Science", "Art"]

# Basic iteration
print("My subjects:")
for subject in subjects:
    print(" -", subject)

print()

# With index — use enumerate()
print("Numbered list:")
for i, subject in enumerate(subjects, start=1):  # start=1 so it starts at 1
    print(f"  {i}. {subject}")
````

````python
# 📖 Iterating over a dictionary
report_card = {
    "Maths": 88,
    "Science": 92,
    "English": 75,
    "Computer Science": 97,
    "Art": 85
}

print("📊 Report Card:")
for subject, score in report_card.items():
    # Add grade letter
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    else:
        grade = "D"
    print(f"  {subject}: {score}/100  →  Grade {grade}")

average = sum(report_card.values()) / len(report_card)
print(f"\nAverage: {average:.1f}")
````

````python
# 🎮 List comprehension — a cooler way to build lists
# Regular way:
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print("Squares (regular):", squares)

# List comprehension — same result, one line!
squares2 = [n ** 2 for n in range(1, 6)]
print("Squares (comprehension):", squares2)

# With a condition — only even squares
even_squares = [n ** 2 for n in range(1, 11) if n % 2 == 0]
print("Even squares:", even_squares)

# Real example: all subjects with score >= 85
report_card = {"Maths": 88, "Science": 92, "English": 75, "CS": 97, "Art": 85}
top_subjects = [subj for subj, score in report_card.items() if score >= 85]
print("Top subjects:", top_subjects)
````

````python
# 🤐 zip() — iterate over two lists at the same time
players = ["Ahmed", "Sara", "Mia", "Omar"]
scores  = [1500,    2300,   1100,  1800]

print("🏆 Leaderboard:")
combined = list(zip(players, scores))
combined.sort(key=lambda x: x[1], reverse=True)  # sort by score

for rank, (player, score) in enumerate(combined, start=1):
    medal = ["🥇", "🥈", "🥉", "  "][rank - 1]
    print(f"  {medal} #{rank} {player}: {score} pts")
````

# 🛠️ Lesson 5 — Functions with Collections

## Why functions?

A **function** is a reusable block of code. When you combine functions with collections, you can build powerful, clean programs.

Think of functions as **tools** in a toolbox — you build them once and use them whenever you need!

```python
def tool_name(input):
    # do something
    return output
```

````python
# 🧰 Functions that work with lists

def get_average(numbers):
    """Calculate the average of a list of numbers."""
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

def get_highest(numbers):
    """Return the highest value."""
    return max(numbers)

def filter_above(numbers, threshold):
    """Return only numbers above the threshold."""
    return [n for n in numbers if n > threshold]

# Test the functions
test_scores = [72, 85, 91, 60, 78, 95, 88, 55]
print("Scores:", test_scores)
print("Average:", get_average(test_scores))
print("Highest:", get_highest(test_scores))
print("Above 80:", filter_above(test_scores, 80))
````

````python
# 📚 Functions that work with dictionaries

def add_student(students_dict, name, age, grade):
    """Add a new student to the dictionary."""
    students_dict[name] = {"age": age, "grade": grade}
    print(f"✅ Added {name} to students")

def find_student(students_dict, name):
    """Find a student and return their info, or a message if not found."""
    if name in students_dict:
        return students_dict[name]
    else:
        return f"❌ Student '{name}' not found"

def list_by_grade(students_dict, grade):
    """Return all students in a specific grade."""
    return [name for name, info in students_dict.items() if info["grade"] == grade]

# Build a student database
students = {}
add_student(students, "Sara", 15, "9th")
add_student(students, "Ahmed", 14, "8th")
add_student(students, "Mia", 15, "9th")
add_student(students, "Omar", 16, "10th")

print("\nLooking up Sara:", find_student(students, "Sara"))
print("Looking up Ali:", find_student(students, "Ali"))
print("9th graders:", list_by_grade(students, "9th"))
````

````python
# 🎮 A simple inventory system using functions

def show_inventory(inventory):
    """Display all items in the inventory."""
    if not inventory:
        print("🎒 Inventory is empty!")
        return
    print("🎒 Your inventory:")
    for item, count in inventory.items():
        print(f"  {item}: x{count}")

def pick_up(inventory, item, count=1):
    """Add item(s) to inventory."""
    inventory[item] = inventory.get(item, 0) + count
    print(f"  ✅ Picked up {count}x {item}")

def use_item(inventory, item, count=1):
    """Use/remove item(s) from inventory."""
    if inventory.get(item, 0) >= count:
        inventory[item] -= count
        if inventory[item] == 0:
            del inventory[item]
        print(f"  ✅ Used {count}x {item}")
    else:
        print(f"  ❌ Not enough {item} in inventory!")

# Play the game!
bag = {}
pick_up(bag, "Health Potion", 3)
pick_up(bag, "Sword")
pick_up(bag, "Shield")
pick_up(bag, "Health Potion", 2)

show_inventory(bag)
print()
use_item(bag, "Health Potion", 4)
use_item(bag, "Bomb")  # doesn't exist!
show_inventory(bag)
````

````python
# 🏋️ EXERCISE 5
# Build a music playlist manager using functions!

def create_playlist(name):
    """Create a new empty playlist."""
    return {"name": name, "songs": []}

def add_song(playlist, song):
    """Add a song to the playlist."""
    playlist["songs"].append(song)
    print(f"🎵 Added '{song}' to {playlist['name']}")

def remove_song(playlist, song):
    """Remove a song if it exists."""
    if song in playlist["songs"]:
        playlist["songs"].remove(song)
        print(f"🗑️ Removed '{song}'")
    else:
        print(f"❌ '{song}' not found in playlist")

def show_playlist(playlist):
    """Display the playlist."""
    print(f"\n🎧 {playlist['name']} ({len(playlist['songs'])} songs):")
    for i, song in enumerate(playlist['songs'], 1):
        print(f"  {i}. {song}")

# Test your playlist manager!
my_mix = create_playlist("My Summer Mix 🌞")
add_song(my_mix, "Blinding Lights")
add_song(my_mix, "Levitating")
add_song(my_mix, "Stay")
add_song(my_mix, "Heat Waves")
show_playlist(my_mix)

print()
remove_song(my_mix, "Stay")
remove_song(my_mix, "Shape of You")  # not in playlist
show_playlist(my_mix)
````

# 🏗️ Lesson 6 — System Architecture: The Full Toolkit

## Putting it all together!

Now we combine **Lists**, **Dictionaries**, **Sets**, **Iteration**, and **Functions** to build a real mini system.

We'll build a **Mini Social App** — like a simple version of Instagram or Twitter! 📱

### System Design
```
📱 Social App
├── 👤 Users (Dictionary: username → profile)
│   ├── name, bio, followers (Set)
│   └── posts (List)
├── 📝 Posts (Dictionary: id → post)
│   ├── content, author, likes (Set)
│   └── tags (Set)
└── 🔧 Functions (create, follow, post, like, search)
```

````python
# 🏗️ Building the Mini Social App

# --- DATA STORAGE ---
users = {}    # dictionary of all users
posts = {}    # dictionary of all posts
post_counter = [0]  # using list so we can modify inside functions

# --- USER FUNCTIONS ---
def create_account(username, name, bio=""):
    """Create a new user account."""
    if username in users:
        print(f"❌ Username '{username}' is already taken!")
        return False
    users[username] = {
        "name": name,
        "bio": bio,
        "followers": set(),   # Set = unique followers
        "following": set(),
        "posts": []           # List = ordered posts
    }
    print(f"✅ Account created for @{username}!")
    return True

def follow(follower, target):
    """Follow another user."""
    if follower not in users or target not in users:
        print("❌ User not found")
        return
    if follower == target:
        print("❌ You can't follow yourself!")
        return
    users[target]["followers"].add(follower)
    users[follower]["following"].add(target)
    print(f"@{follower} is now following @{target}")

def show_profile(username):
    """Display a user's profile."""
    if username not in users:
        print(f"❌ @{username} not found")
        return
    u = users[username]
    print(f"\n👤 @{username}")
    print(f"   {u['name']}")
    if u['bio']:
        print(f"   {u['bio']}")
    print(f"   📊 {len(u['followers'])} followers · {len(u['following'])} following · {len(u['posts'])} posts")

print("Social App initialized! 🚀")
````

````python
# --- POST FUNCTIONS ---

def create_post(username, content, tags=None):
    """Create a new post."""
    if username not in users:
        print("❌ User not found")
        return
    post_counter[0] += 1
    post_id = f"post_{post_counter[0]}"
    posts[post_id] = {
        "author": username,
        "content": content,
        "likes": set(),    # Set = each user can like once!
        "tags": set(tags) if tags else set()
    }
    users[username]["posts"].append(post_id)
    print(f"📝 @{username} posted: \"{content[:40]}...\"" if len(content) > 40 else f"📝 @{username} posted: \"{content}\"")
    return post_id

def like_post(username, post_id):
    """Like or unlike a post (toggle)."""
    if post_id not in posts:
        print("❌ Post not found")
        return
    if username in posts[post_id]["likes"]:
        posts[post_id]["likes"].discard(username)
        print(f"💔 @{username} unliked the post")
    else:
        posts[post_id]["likes"].add(username)
        print(f"❤️  @{username} liked the post ({len(posts[post_id]['likes'])} likes total)")

def search_by_tag(tag):
    """Find all posts with a specific tag."""
    results = [(pid, p) for pid, p in posts.items() if tag in p["tags"]]
    print(f"\n🔍 Posts tagged '#{tag}':")
    if not results:
        print("  No posts found")
    for pid, p in results:
        print(f"  @{p['author']}: \"{p['content']}\" ({len(p['likes'])} ❤️)")

def show_feed(username):
    """Show posts from people the user follows."""
    following = users[username]["following"]
    print(f"\n📰 {username}'s Feed:")
    feed_posts = [(pid, p) for pid, p in posts.items() if p["author"] in following]
    if not feed_posts:
        print("  Nothing to show yet. Follow some people!")
    for pid, p in feed_posts:
        likes = len(p['likes'])
        print(f"  @{p['author']}: \"{p['content']}\" — {likes} ❤️")

print("Post functions ready! ✅")
````

````python
# 🎬 DEMO — Run the social app!

# Create accounts
create_account("sara_codes", "Sara Ahmed", "💻 Aspiring developer | 📚 Book lover")
create_account("ahmed_plays", "Ahmed Ali", "🎮 Gamer | ⚽ Football fan")
create_account("mia_art", "Mia Hassan", "🎨 Artist | ✨ Creative soul")
create_account("omar_tech", "Omar Karim", "🤖 Tech enthusiast | 🚀 Future engineer")

print()

# Follow each other
follow("sara_codes", "ahmed_plays")
follow("sara_codes", "mia_art")
follow("ahmed_plays", "sara_codes")
follow("mia_art", "sara_codes")
follow("omar_tech", "sara_codes")

print()

# Create posts
p1 = create_post("sara_codes", "Just built my first Python app! 🐍", ["python", "coding", "beginner"])
p2 = create_post("ahmed_plays", "FIFA tournament tonight! Who's joining? 🎮", ["gaming", "fifa", "sports"])
p3 = create_post("mia_art", "New digital artwork finished! Link in bio 🎨", ["art", "digitalart", "creative"])
p4 = create_post("sara_codes", "Python tip: use list comprehension! [x for x in range(10)]", ["python", "tips", "coding"])
p5 = create_post("omar_tech", "AI is everywhere now. Learn to code! 🤖", ["tech", "ai", "coding"])

print()

# Like posts
like_post("ahmed_plays", p1)
like_post("mia_art", p1)
like_post("omar_tech", p1)
like_post("sara_codes", p2)
like_post("sara_codes", p3)
like_post("mia_art", p4)
like_post("ahmed_plays", p4)
````

````python
# 🔎 Explore the app
show_profile("sara_codes")
show_profile("ahmed_plays")

print()
show_feed("sara_codes")

print()
search_by_tag("coding")
search_by_tag("gaming")
````

````python
# 📊 Analytics — using all collection types together!

def app_summary():
    """Show overall app statistics."""
    print("\n" + "="*40)
    print("📱 APP ANALYTICS DASHBOARD")
    print("="*40)

    # Basic stats
    total_likes = sum(len(p["likes"]) for p in posts.values())
    all_tags = set(tag for p in posts.values() for tag in p["tags"])

    print(f"👥 Total users: {len(users)}")
    print(f"📝 Total posts: {len(posts)}")
    print(f"❤️  Total likes: {total_likes}")
    print(f"🏷️  Unique tags used: {len(all_tags)}")
    print(f"   Tags: {', '.join(sorted(all_tags))}")

    # Most liked post
    most_liked_id = max(posts, key=lambda pid: len(posts[pid]["likes"]))
    most_liked = posts[most_liked_id]
    print(f"\n🏆 Most liked post: @{most_liked['author']}")
    print(f"   \"{most_liked['content']}\"")
    print(f"   {len(most_liked['likes'])} likes")

    # Most followed user
    most_followed = max(users, key=lambda u: len(users[u]["followers"]))
    print(f"\n⭐ Most followed: @{most_followed} with {len(users[most_followed]['followers'])} followers")

    # User leaderboard
    print("\n📊 Follower Leaderboard:")
    sorted_users = sorted(users.items(), key=lambda u: len(u[1]["followers"]), reverse=True)
    for i, (uname, info) in enumerate(sorted_users, 1):
        print(f"  {i}. @{uname} — {len(info['followers'])} followers, {len(info['posts'])} posts")

app_summary()
````

````python
# 🏋️ FINAL CHALLENGE — Extend the Social App!
# Add these 3 new features to the app:

# FEATURE 1: add_comment(post_id, username, comment)
# - Add a "comments" list to posts when they're created (modify create_post)
# - Store each comment as a dict: {"user": username, "text": comment}
# - Print the comment after adding

def add_comment(post_id, username, comment):
    if post_id not in posts:
        print("❌ Post not found")
        return
    # Add comments list if it doesn't exist (for older posts)
    if "comments" not in posts[post_id]:
        posts[post_id]["comments"] = []
    posts[post_id]["comments"].append({"user": username, "text": comment})
    print(f"💬 @{username}: {comment}")

# Test Feature 1:
print("=== Testing Comments ===")
add_comment(p1, "ahmed_plays", "Wow amazing! Teach me! 🙏")
add_comment(p1, "omar_tech", "Python is the best for beginners!")
add_comment(p1, "mia_art", "I want to learn coding too 😊")
print(f"\nComments on post 1: {len(posts[p1]['comments'])}")

print()

# FEATURE 2: suggest_friends(username)
# - Find users who are followed by the people you follow (friends of friends)
# - Exclude yourself and people you already follow
# - Return as a sorted list

def suggest_friends(username):
    following = users[username]["following"]
    suggestions = set()
    for followed_user in following:
        for their_follow in users[followed_user]["following"]:
            if their_follow != username and their_follow not in following:
                suggestions.add(their_follow)
    print(f"\n💡 Friend suggestions for @{username}:")
    for suggestion in sorted(suggestions):
        print(f"  👤 @{suggestion}")

# Test Feature 2:
print("=== Friend Suggestions ===")
suggest_friends("sara_codes")
suggest_friends("ahmed_plays")
````

# 📊 Lesson 1 — Data vs Information

- **Data** → raw numbers or text with **no context** (just values)
- **Information** → data **+ context + meaning** (actionable)
- **Data Science** turns raw data into knowledge using statistics + code

````python
# Raw data — just numbers, no meaning yet
raw_temps = [36.6, 37.1, 36.8, 38.2, 36.5]
print("Raw data:", raw_temps)

# Information — data + context + decision
patients = [
    {"name": "Sara",  "temp_c": 36.6},
    {"name": "Omar",  "temp_c": 37.1},
    {"name": "Layla", "temp_c": 36.8},
    {"name": "Ahmed", "temp_c": 38.2},
    {"name": "Nour",  "temp_c": 36.5},
]

print("\nInformation (data + context):")
for p in patients:
    status = "🌡️  FEVER — monitor!" if p["temp_c"] >= 38 else "✅ Normal"
    print(f"  {p['name']:6} → {p['temp_c']}°C  {status}")
````

````python
# Guess vs data-driven prediction
import random

# Team's win rate from the last 20 matches
results = ["W", "W", "L", "W", "W", "L", "W", "W", "W", "L",
           "W", "L", "W", "W", "W", "L", "W", "W", "L", "W"]

wins  = results.count("W")
total = len(results)
win_rate = wins / total

print(f"Last {total} matches: {wins} wins")
print(f"Win rate: {win_rate:.0%}")
print()
print("Gut feeling prediction: 'I think they'll win'  ← no evidence")
print(f"Data-driven prediction:  {win_rate:.0%} chance of winning  ← based on history")
````

# 🏷️ Lesson 2 — Data Types: Numerical vs Categorical

| Type | Description | Example |
|------|-------------|---------|
| **Numerical** | Numbers you can add, subtract, average | Scores, salary, height |
| **Categorical** | Labels / groups — no arithmetic | Colour, subject, country |

````python
def classify_type(value):
    """Returns 'numerical' or 'categorical'."""
    try:
        float(value)
        return "📊 Numerical"
    except (ValueError, TypeError):
        return "🏷️  Categorical"

samples = [
    ("42",       "exam score"),
    ("red",      "favourite colour"),
    ("1.75",     "height in metres"),
    ("football", "favourite sport"),
    ("98.6",     "body temperature"),
    ("Egypt",    "country"),
]

print("Data type classifier:")
for val, desc in samples:
    print(f"  {classify_type(val):16} ← {val!r:12}  ({desc})")
````

# 📐 Lesson 3 — Summary Statistics: Mean, Median, Mode

| Measure | What it is | When to use |
|---------|------------|--------------|
| **Mean** | sum ÷ count | Normal data, no extreme values |
| **Median** | middle value (sorted) | Data with outliers |
| **Mode** | most frequent value | Categorical data |

````python
# Mean, median, mode — built from scratch (no libraries)

def mean(values):
    return sum(values) / len(values)

def median(values):
    s = sorted(values)
    n = len(s)
    mid = n // 2
    if n % 2 == 1:
        return s[mid]
    return (s[mid - 1] + s[mid]) / 2

def mode(values):
    counts = {}
    for v in values:
        counts[v] = counts.get(v, 0) + 1
    return max(counts, key=counts.get)

scores = [72, 85, 91, 68, 85, 79, 85, 62, 95, 79]
print(f"Scores: {scores}")
print(f"Mean:   {mean(scores):.1f}")
print(f"Median: {median(scores)}")
print(f"Mode:   {mode(scores)}  ← appears most often")
````

````python
# Outliers pull the mean — median is safer
normal_salaries  = [3000, 3200, 2800, 3100, 3050]
ceo_added        = [3000, 3200, 2800, 3100, 3050, 500000]  # CEO joins

print("Without CEO:")
print(f"  Mean: {mean(normal_salaries):,.0f}  |  Median: {median(normal_salaries):,.0f}")

print("\nWith CEO (outlier):")
print(f"  Mean: {mean(ceo_added):,.0f}  ← dragged up!")
print(f"  Median: {median(ceo_added):,.0f}  ← barely moves")
print("\n💡 When outliers exist → use Median, not Mean")
````

# 🧹 Lesson 4 — Data Cleaning & GIGO

> **GIGO — Garbage In, Garbage Out**  
> 60-80% of a data scientist's time is spent cleaning data.

Common dirty data: `None`, impossible values, duplicates, wrong types.

````python
# Messy temperature data — spot the problems!
raw_temps = [36.6, None, 37.1, -99, 36.8, 38.2, "N/A", 36.5, 150, None]
print("Original:", raw_temps)

def clean_temps(raw):
    cleaned  = []
    rejected = []
    for val in raw:
        if val is None or not isinstance(val, (int, float)):
            rejected.append(val)
            continue
        if val < 35 or val > 42:          # impossible body temperature
            rejected.append(val)
            continue
        cleaned.append(val)
    return cleaned, rejected

clean, rejected = clean_temps(raw_temps)
print(f"Cleaned:  {clean}")
print(f"Rejected: {rejected}")

# How much does dirty data distort the mean?
all_numbers = [v for v in raw_temps if isinstance(v, (int, float)) and v is not None]
print(f"\nMean BEFORE cleaning: {mean(all_numbers):.1f}°C  ← wrong!")
print(f"Mean AFTER  cleaning: {mean(clean):.1f}°C  ← correct")
````

# 🔁 Lesson 5 — The 5-Step Analysis Process

```
Ask  →  Collect  →  Clean  →  Analyse  →  Share
```

| Step | What you do |
|------|-------------|
| **Ask** | Define the question clearly |
| **Collect** | Gather relevant data |
| **Clean** | Fix errors, remove garbage |
| **Analyse** | Find patterns, calculate measures |
| **Share** | Visualise and communicate the result |

````python
# Example: Which gym class is most popular?

example_plan = {
    "1_Ask":     "Which gym class gets the most sign-ups?",
    "2_Collect": "Attendance records for all classes — last 3 months",
    "3_Clean":   "Remove duplicate entries and fix missing dates",
    "4_Analyse": "Count sign-ups per class and find the top 3",
    "5_Share":   "Bar chart + one-sentence recommendation to management",
}

print("5-Step Analysis Plan — Gym Example:")
for step, desc in example_plan.items():
    print(f"  {step}: {desc}")
````

````python
# Step 4 in action: find the top gym classes

attendance = {
    "Yoga":      [22, 25, 23, 24, 26, 22, 25, 27, 24, 23, 25, 24],
    "Spinning":  [30, 28, 31, 29, 32, 30, 31, 33, 29, 30, 32, 31],
    "Zumba":     [18, 20, 19, 21, 17, 20, 19, 22, 18, 20, 21, 19],
    "Pilates":   [15, 14, 16, 15, 13, 16, 14, 15, 16, 14, 15, 16],
    "Boxing":    [27, 26, 28, 25, 29, 27, 28, 30, 26, 27, 29, 28],
}

print("Average monthly attendance:")
averages = {}
for cls, nums in attendance.items():
    avg = mean(nums)
    averages[cls] = avg
    bar = "█" * int(avg)
    print(f"  {cls:10} {avg:5.1f}  {bar}")

top_class = max(averages, key=averages.get)
print(f"\n📊 Result: '{top_class}' is the most popular class with {averages[top_class]:.0f} avg sign-ups.")
````

## ⚠️ Correlation Trap — Correlation ≠ Causation

Two things rising together does **not** mean one causes the other.  
There is often a **hidden third factor**.

````python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
ice_cream = [100, 110, 200, 350,  600,  900, 1200, 1100,  700, 350, 150, 120]
sunburn   = [  5,   8,  20,  45,  100,  180,  250,  220,  120,  40,  10,   6]
temp_c    = [ 10,  12,  18,  23,   28,   33,   38,   37,   30,  22,  14,  11]

print(f"{'Month':5} | {'Ice cream':10} | {'Sunburn':8} | Temp")
print("-" * 38)
for m, ic, sb, t in zip(months, ice_cream, sunburn, temp_c):
    bar = "🌡️" * (t // 10)
    print(f"{m:5} | {ic:10} | {sb:8} | {t}°C {bar}")

print("\n⚠️  Ice cream sales and sunburn cases both rise in summer.")
print("   Does eating ice cream CAUSE sunburn?  ❌  NO!")
print("   Hidden factor: HOT WEATHER causes BOTH.")
print("   Correlation ≠ Causation")
````
